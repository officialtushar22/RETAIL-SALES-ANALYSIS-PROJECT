silver layer transformation

Transaction Table

```python
from pyspark.sql.functions import col


df_transaction_clean = (
    df_transaction
    .filter(col("transaction_id").isNotNull())
    .filter(col("quantity") > 0)
    #.filter(col("price") > 0)
)

```

Product table

```python
df_product_clean = (
    df_product
    .filter(col("product_id").isNotNull())
    .filter(col("price") > 0)
    .dropDuplicates(["product_id"])
)
```

Store Table

```python
df_store_clean = (
    df_store
    .filter(col("store_id").isNotNull())
    .dropDuplicates(["store_id"])
)
```

Customer Table

```python
df_customer_clean = (
    df_customer
    .filter(col("customer_id").isNotNull())
    .dropDuplicates(["customer_id"])
)
```

Join All Sources

```python
df_silver = (
    df_transaction_clean
    .join(df_product_clean, "product_id", "left")
    .join(df_store_clean, "store_id", "left")
    .join(df_customer_clean, "customer_id", "left")
)
```

Feature Engineering

```python
from pyspark.sql.functions import year, month, concat_ws

df_silver = (
    df_silver
    .withColumn("total_amount", col("quantity") * col("price"))
    .withColumn("order_year", year(col("transaction_date")))
    .withColumn("order_month", month(col("transaction_date")))
    .withColumn("customer_name", concat_ws(" ", col("first_name"), col("last_name")))
)
```

```python
df_silver.write \
    .format("parquet") \
    .mode("overwrite") \
    .save("abfss://retail@retailrsa.dfs.core.windows.net/silver/new_silver/")
```

Gold Transformation

Create Discount Buckets (Campaign Types)

```python
from pyspark.sql.functions import when, col

df_campaign = (
    df_silver
    .withColumn(
        "campaign_type",
        when(col("price") < 50, "High Discount")
        .when((col("price") >= 50) & (col("price") < 200), "Medium Discount")
        .otherwise("Low / No Discount")
    )
)
```

Campaign Performance Metrics

```python
from pyspark.sql.functions import sum, avg, count

df_campaign_perf = (
    df_campaign
    .groupBy("campaign_type", "category")
    .agg(
        count("transaction_id").alias("total_transactions"),
        sum("quantity").alias("total_units_sold"),
        sum("total_amount").alias("total_revenue"),
        avg("total_amount").alias("avg_order_value")
    )
)
```

Classify Campaign Success

```python
df_campaign_result = (
    df_campaign_perf
    .withColumn(
        "campaign_status",
        when(col("total_revenue") > 50000, "Successful")
        .otherwise("Unsuccessful")
    )
)
```

Write Campaign Analysis to GOLD (Delta)

```python
df_campaign_result.write \
    .format("parquet") \
    .mode("overwrite") \
    .save("abfss://retail@retailrsa.dfs.core.windows.net/gold/new_gold_compaign/")
```

GOLD LAYER â€“ ANALYTICS TRANSFORMATIONS

Product sales

```python
from pyspark.sql.functions import sum

df_gold_product = (
    df_silver
    .groupBy("product_id", "product_name", "category")
    .agg(
        sum("quantity").alias("total_quantity_sold"),
        sum("total_amount").alias("total_sales")
    )
)
```

Store sales

```python
df_gold_store = (
    df_silver
    .groupBy("store_id", "store_name", "location")
    .agg(
        sum("total_amount").alias("store_sales")
    )
)
```

Customer Lifetime Value

```python
df_gold_customer = (
    df_silver
    .groupBy("customer_id", "customer_name", "city")
    .agg(
        sum("total_amount").alias("customer_lifetime_value"),
        sum("quantity").alias("total_items_purchased")
    )
)
```

Monthly Sales Trend

```python
df_gold_monthly = (
    df_silver
    .groupBy("order_year", "order_month")
    .agg(
        sum("total_amount").alias("monthly_sales")
    )
    .orderBy("order_year", "order_month")
)
```
